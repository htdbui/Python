<h1 style="font-size:60px;font-weight:bold;">Assertion Functions</h1>

__testing.assert_frame_equal(left, right[, ...])__

```
from pandas.testing import assert_frame_equal

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
assert_frame_equal(df1, df2)
```

__testing.assert_series_equal(left, right[, ...])__

```
from pandas.testing import assert_series_equal

s1 = pd.Series([1, 2, 3])
s2 = pd.Series([1, 2, 3])
assert_series_equal(s1, s2)
```

__testing.assert_index_equal(left, right[, ...])__

```
from pandas.testing import assert_index_equal

i1 = pd.Index(['a', 'b', 'c'])
i2 = pd.Index(['a', 'b', 'c'])
assert_index_equal(i1, i2)
```

__testing.assert_extension_array_equal(left, right)__

```
from pandas.testing import assert_extension_array_equal

ea1 = pd.arrays.IntegerArray([1, 2, 3])
ea2 = pd.arrays.IntegerArray([1, 2, 3])
assert_extension_array_equal(ea1, ea2)
```
