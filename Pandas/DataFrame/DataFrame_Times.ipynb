<h1 style="font-size:60px;font-weight:bold;">Time Series-related</h1>

__DataFrame.asfreq(freq[, method, how, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D'))
df1.asfreq('2D')
```

__DataFrame.asof(where[, subset])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D'))
df1.asof('2022-01-04')
```

__DataFrame.shift([periods, freq, axis, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.shift(periods=2)
```

__DataFrame.first_valid_index()__
```
df1 = pd.DataFrame({'a': [1, 2, np.nan, 4, np.nan, 6]})
df1.first_valid_index()
```

__DataFrame.last_valid_index()__
```
df1 = pd.DataFrame({'a': [1, 2, np.nan, 4, np.nan, 6]})
df1.last_valid_index()
```

__DataFrame.resample(rule[, axis, closed, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D'))
df1.resample('2D').mean()
```

__DataFrame.to_period([freq, axis, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D'))
df1.to_period('M')
```

__DataFrame.to_timestamp([freq, how, axis, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='M'))
df1.to_timestamp(freq='D')
```

__DataFrame.tz_convert(tz[, axis, level, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D', tz='UTC'))
df1.tz_convert('US/Eastern')
```

__DataFrame.tz_localize(tz[, axis, level, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]}, index=pd.date_range(start='2022-01-01', periods=6, freq='D'))
df1.tz_localize('UTC')
```
