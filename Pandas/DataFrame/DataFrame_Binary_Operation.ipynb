__DataFrame.combine(other, func[, fill_value, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df2 = pd.DataFrame({'a': [10, 20, 30], 'b': [40, 50, 60]})
df1.combine(df2, lambda x, y: x if x>y else y)
```
