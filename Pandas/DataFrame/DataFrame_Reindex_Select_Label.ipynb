<h1 style="font-size:60px;font-weight:bold;">Reindexing / Selection / Label Manipulation</h1>


__DataFrame.add_prefix(prefix[, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.add_prefix('x_')
```

__DataFrame.add_suffix(suffix[, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.add_suffix('_y')
```

__DataFrame.align(other[, join, axis, level, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df2 = pd.DataFrame({'a': [2, 3, 4], 'c': [5, 6, 7]})
df1, df2 = df1.align(df2, join='outer', axis=1)
```

__DataFrame.at_time(time[, asof, axis])__
```
df1 = pd.DataFrame({'time': ['2022-01-01 12:00:00', '2022-01-01 13:00:00', '2022-01-01 14:00:00'], 'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.set_index('time', inplace=True)
df1.at_time('13:00:00')
```

__DataFrame.between_time(start_time, end_time)__
```
df1 = pd.DataFrame({'time': ['2022-01-01 12:00:00', '2022-01-01 13:00:00', '2022-01-01 14:00:00'], 'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.set_index('time', inplace=True)
df1.between_time('12:30:00', '13:30:00')
```

__DataFrame.drop([labels, axis, index, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.drop(columns=['b'])
```

__DataFrame.drop_duplicates([subset, keep, ...])__
```
df1 = pd.DataFrame({'a': [1, 1, 2], 'b': [3, 3, 4]})
df1.drop_duplicates()
```

__DataFrame.duplicated([subset, keep])__
```
df1 = pd.DataFrame({'a': [1, 1, 2], 'b': [3, 3, 4]})
df1.duplicated()
```

__DataFrame.equals(other)__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df2 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.equals(df2)
```

__DataFrame.filter([items, like, regex, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]})
df1.filter(regex='^a|c$', axis=1)
```

__DataFrame.first(offset)__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.first('2D')
```

__DataFrame.head([n])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.head(2)
```

__DataFrame.idxmax([axis, skipna, numeric_only])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.idxmax()
```

__DataFrame.idxmin([axis, skipna, numeric_only])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.idxmin()
```

__DataFrame.last(offset)__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.last('2D')
```

__DataFrame.reindex([labels, index, columns, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.reindex(index=[0, 2, 1])
```

__DataFrame.reindex_like(other[, method, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df2 = pd.DataFrame({'b': [6, 5, 4], 'a': [3, 2, 1]})
df1.reindex_like(df2)
```

__DataFrame.rename([mapper, index, columns, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.rename(columns={'a': 'x', 'b': 'y'})
```

__DataFrame.rename_axis([mapper, index, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.rename_axis('id')
```

__DataFrame.reset_index([level, drop, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.reset_index(drop=True)
```

__DataFrame.sample([n, frac, replace, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.sample(n=2)
```

__DataFrame.set_axis(labels, *[, axis, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.set_axis(['x', 'y'], axis=1)
```

__DataFrame.set_index(keys, *[, drop, append, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.set_index('a')
```

__DataFrame.tail([n])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.tail(2)
```

__DataFrame.take(indices[, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.take([2, 0], axis=0)
```

__DataFrame.truncate([before, after, axis, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.truncate(before=2)
```
