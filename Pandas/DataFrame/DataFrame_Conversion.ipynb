<h1 style="font-size:60px;font-weight:bold;">Serialization / IO / Conversion</h1>

__DataFrame.from_dict(data[, orient, dtype, ...])__
```
data = {'col1': [1, 2], 'col2': [3, 4]}
df = pd.DataFrame.from_dict(data, orient='columns')
```

__DataFrame.from_records(data[, index, ...])__
```
data = [(1, 'A', 100), (2, 'B', 200), (3, 'C', 300)]
df = pd.DataFrame.from_records(data, columns=['col1', 'col2', 'col3'], index='col1')
```

__DataFrame.to_orc([path, engine, index, ...])__
```
df.to_orc('example.orc')
```

__DataFrame.to_parquet([path, engine, ...])__
```
df.to_parquet('example.parquet')
```

__DataFrame.to_pickle(path[, compression, ...])__
```
df.to_pickle('example.pkl')
```

__DataFrame.to_csv([path_or_buf, sep, na_rep, ...])__
```
df.to_csv('example.csv', sep=',', na_rep='NA')
```

__DataFrame.to_hdf(path_or_buf, key[, mode, ...])__
```
df.to_hdf('example.h5', key='df', mode='w')
```

__DataFrame.to_sql(name, con[, schema, ...])__
```
from sqlalchemy import create_engine
engine = create_engine('sqlite:///example.db')
df.to_sql('table_name', con=engine)
```

__DataFrame.to_dict([orient, into, index])__
```
df.to_dict(orient='list')
```

__DataFrame.to_excel(excel_writer[, ...])__
```
df.to_excel('example.xlsx', sheet_name='Sheet1', index=False)
```

__DataFrame.to_json([path_or_buf, orient, ...])__
```
df.to_json('example.json', orient='records', lines=True)
```

__DataFrame.to_html([buf, columns, col_space, ...])__
```
df.to_html(buf='example.html', columns=['col1', 'col2'], col_space=10)
```

__DataFrame.to_feather(path, **kwargs)__
```
df.to_feather('example.feather')
```

__DataFrame.to_latex([buf, columns, header, ...])__
```
df.to_latex(buf='example.tex', columns=['col1', 'col2'], header=['Column1', 'Column2'])
```

__DataFrame.to_stata(path, *[, convert_dates, ...])__
```
df.to_stata('example.dta', version=117)
```

__DataFrame.to_gbq(destination_table[, ...])__
```
from google.oauth2 import service_account
credentials = service_account.Credentials.from_service_account_file('path/to/credentials.json')
df.to_gbq(destination_table='my_dataset.my_table', project_id='my_project_id', credentials=credentials, if_exists='replace')
```

__DataFrame.to_records([index, column_dtypes, ...])__
```
df.to_records(index=False)
```

__DataFrame.to_string([buf, columns, ...])__
```
df.to_string(buf='example.txt', columns=['col1', 'col2'], index=False)
```

__DataFrame.to_clipboard([excel, sep])__
```
df.to_clipboard(excel=True, sep=',')
```

__DataFrame.to_markdown([buf, mode, index, ...])__
```
df.to_markdown(buf='example.md', mode='table', index=False)
```

__DataFrame.style__
```
df.style.set_properties(**{'text-align': 'center', 'width': '150px'})
```

__DataFrame.__dataframe__([nan_as_null, ...])__
```
df.__dataframe__()
```
