<h1 style="font-size:60px;font-weight:bold;">Reshaping, Sorting, Transposing</h1>

__DataFrame.droplevel(level[, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]}, index=[['x', 'x', 'y'], [0, 1, 0]])
df1.droplevel(1)
```

__DataFrame.pivot(*, columns[, index, values])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'], 'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'], 'C': [1, 2, 3, 4, 5, 6, 7, 8], 'D': [2, 4, 5, 6, 7, 8, 9, 10]})
df1.pivot(columns='A', values='D')
```

__DataFrame.pivot_table([values, index, ...])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'], 'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'], 'C': [1, 2, 3, 4, 5, 6, 7, 8], 'D': [2, 4, 5, 6, 7, 8, 9, 10]})
df1.pivot_table(values='D', index='B', columns='A', aggfunc=np.sum)
```

__DataFrame.reorder_levels(order[, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]}, index=[['x', 'x', 'y'], [0, 1, 0]])
df1.reorder_levels([1, 0])
```

__DataFrame.sort_values(by, *[, axis, ...])__
```
df1 = pd.DataFrame({'a': [3, 2, 1], 'b': [6, 5, 4]})
df1.sort_values('a')
```

__DataFrame.sort_index(*[, axis, level, ...])__
```
df1 = pd.DataFrame({'a': [3, 2, 1], 'b': [6, 5, 4]}, index=['c', 'a', 'b'])
df1.sort_index()
```

__DataFrame.nlargest(n, columns[, keep])__
```
df1 = pd.DataFrame({'a': [3, 2, 1], 'b': [6, 5, 4]})
df1.nlargest(2, columns='a')
```

__DataFrame.nsmallest(n, columns[, keep])__
```
df1 = pd.DataFrame({'a': [3, 2, 1], 'b': [6, 5, 4]})
df1.nsmallest(2, columns='a')
```

__DataFrame.swaplevel([i, j, axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]}, index=[['x', 'x', 'y'], [0, 1, 0]])
df1.swaplevel(0, 1)
```

__DataFrame.stack([level, dropna])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'], 'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'], 'C': [1, 2, 3, 4, 5, 6, 7, 8], 'D': [2, 4, 5, 6, 7, 8, 9, 10]})
df1.stack()
```

__DataFrame.unstack([level, fill_value])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'], 'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'], 'C': [1, 2, 3, 4, 5, 6, 7, 8], 'D': [2, 4, 5, 6, 7, 8, 9, 10]})
df1.set_index(['A', 'B']).unstack()
```

__DataFrame.swapaxes(axis1, axis2[, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.swapaxes(0, 1)
```

__DataFrame.melt([id_vars, value_vars, ...])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'foo', 'bar', 'foo', 'bar', 'foo', 'foo'], 'B': ['one', 'one', 'two', 'three', 'two', 'two', 'one', 'three'], 'C': [1, 2, 3, 4, 5, 6, 7, 8], 'D': [2, 4, 5, 6, 7, 8, 9, 10]})
df1.melt(id_vars=['A', 'B'], value_vars=['C', 'D'], var_name='variable', value_name='value')
```

__DataFrame.explode(column[, ignore_index])__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'baz'], 'B': [[1, 2], [3, 4, 5], [6]]})
df1.explode('B')
```

__DataFrame.squeeze([axis])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df2 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6], 'c': [7, 8, 9]})
df1.squeeze()
```

__DataFrame.to_xarray()__
```
df1 = pd.DataFrame({'A': ['foo', 'bar', 'baz'], 'B': [1, 2, 3], 'C': [4, 5, 6]})
df1.to_xarray()
```

__DataFrame.T__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.T
```

__DataFrame.transpose(*args[, copy])__
```
df1 = pd.DataFrame({'a': [1, 2, 3], 'b': [4, 5, 6]})
df1.transpose()
```
