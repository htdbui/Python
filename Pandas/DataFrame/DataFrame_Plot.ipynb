<h1 style="font-size:60px;font-weight:bold;">Plotting</h1>

__DataFrame.plot([x, y, kind, ax, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6], 'b': [2, 4, 1, 5, 7, 2]})
df1.plot(x='a', y='b', kind='line')
```

__DataFrame.plot.area([x, y, stacked])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6], 'b': [2, 4, 1, 5, 7, 2]})
df1.plot(kind='area')
```

__DataFrame.plot.bar([x, y])__
```
df1 = pd.DataFrame({'a': ['A', 'B', 'C', 'D'], 'b': [2, 4, 1, 5]})
df1.plot(kind='bar', x='a', y='b')
```

__DataFrame.plot.barh([x, y])__
```
df1 = pd.DataFrame({'a': ['A', 'B', 'C', 'D'], 'b': [2, 4, 1, 5]})
df1.plot(kind='barh', x='a', y='b')
```

__DataFrame.plot.box([by])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.plot(kind='box')
```

__DataFrame.plot.density([bw_method, ind])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.plot(kind='density')
```

__DataFrame.plot.hexbin(x, y[, C, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6], 'b': [2, 4, 1, 5, 7, 2]})
df1.plot(kind='hexbin', x='a', y='b', gridsize=10)
```

__DataFrame.plot.hist([by, bins])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.plot(kind='hist')
```

__DataFrame.plot.kde([bw_method, ind])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.plot(kind='kde')
```

__DataFrame.plot.line([x, y])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6], 'b': [2, 4, 1, 5, 7, 2]})
df1.plot(x='a', y='b', kind='line')
```

__DataFrame.plot.pie(**kwargs)__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4]})
df1.plot(kind='pie', y='a')
```

__DataFrame.plot.scatter(x, y[, s, c])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6], 'b': [2, 4, 1, 5, 7, 2]})
df1.plot(kind='scatter', x='a', y='b')
```

__DataFrame.boxplot([column, by, ax, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.boxplot()
```

__DataFrame.hist([column, by, grid, ...])__
```
df1 = pd.DataFrame({'a': [1, 2, 3, 4, 5, 6]})
df1.hist()
```
