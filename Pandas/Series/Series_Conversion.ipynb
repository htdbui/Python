<h1 style="font-size:60px;font-weight:bold;">Serialization / IO / conversion</h1>

__Series.to_pickle(path[, compression, ...])__
```
s = pd.Series([1, 2, 3])
s.to_pickle('data.pkl')
```

__Series.to_csv([path_or_buf, sep, na_rep, ...])__
```
s = pd.Series([1, 2, 3])
s.to_csv('data.csv')
```

__Series.to_dict([into])__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.to_dict()
```

__Series.to_excel(excel_writer[, sheet_name, ...])__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
writer = pd.ExcelWriter('data.xlsx')
s.to_excel(writer)
writer.save()
```

__Series.to_frame([name])__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
df = s.to_frame()
```

__Series.to_xarray()__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
x = s.to_xarray()
```

__Series.to_hdf(path_or_buf, key[, mode, ...])__
```
s = pd.Series([1, 2, 3])
s.to_hdf('data.h5', key='s')
```

__Series.to_sql(name, con[, schema, ...])__
```
import sqlite3
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
con = sqlite3.connect('mydatabase.db')
s.to_sql('mytable', con)
```
