<h1 style="font-size:60px;font-weight:bold;">Creat a Series</h1>

    Syntax: Series([data, index, dtype, name, copy, ...])
```
pd.Series(range(3))
pd.Series(enumerate('abc', 1))
pd.Series(zip('abc','def',[1,2,3]))
```

<h1 style="font-size:60px;font-weight:bold;">Attributes</h1>

```
s = pd.Series([1, 2, 3, 4, 5])
s.index
s.array # return Pandas array
s.values # return NumPy array
s.dtype
s.dtypes # return data types as a pandas series
s.name = 'MySeries'; s.name
s.T # return transpose
s.size # return number of elements
s.shape
s.ndim # return number of dimensions
s.hasnans # return True if it has NaN, False otherwise
s.empty # return True if the Series is empty, False otherwise
```
```
s.nbytes # return number of bytes used
```
__Series.memory_usage([index, deep])__: 
```
s.memory_usage(index=False) # return the memory usage of the Series
```
__Series.flags__: returns a dictionary of the flags
```
s.flags
```
__Series.set_flags(*\[, copy, ...\])__: sets the flags
```
s.set_flags(write=False)
```

<h1 style="font-size:60px;font-weight:bold;">Conversion</h1>

```
s = pd.Series([1, 2, '3', 4.0, '5'])
```
__Series.astype(dtype\[, copy, errors\])__: return a new series
```
s_int = s.astype(int)
s_int[0] = 0; s
```
__Series.copy(\[deep\])__: create a copy of the Series
```
s_copy = s.copy(deep=True)
```
__Series.to_numpy(\[dtype, copy, na_value\])__: convert the Series to a NumPy array, return a new ndarray
```
s.to_numpy()
```
__Series.\_\_array\_\_(\[dtype])__: convert the Series to a NumPy array, return only the view
    
    s.__array__()
    
__Series.to_list()__: convert to a Python list
```
s.to_list()
```
__Series.to_period(\[freq, copy\])__: Convert Series from DatetimeIndex to PeriodIndex

__Series.to_timestamp(\[freq, how, copy\])__: Cast to DatetimeIndex of Timestamps, at beginning of period

<h1 style="font-size:60px;font-weight:bold;">Indexing</h1>

```
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
```
__Series.get(key\[, default\])__
```
s.get('b')
s.get('f', 'No such key')
```
__Series.at()__
```
s.at['c']
```
__Series.iat()__
```
s.iat[2]
```
__Series.loc()__
```
s.loc[['a', 'c', 'e']]
```
__Series.iloc()__
```
s.iloc[[0, 2, 4]]
```
__Series.\_\_iter\_\_()__: return an iterator
```
s_iter = s.__iter__()
for e in s_iter:
    print(e)
```
__Series.items()__: iterate over (index, value) pairs 
```
for index, value in s.items():
    print(index, value)
```
__Series.keys()__: return an Index of the keys
```
s.keys()
```
__Series.pop()__
```
s.pop('b')
```
__Note__: <br>
- Series.at() is very fast to access a single value by label(key)
- Series.get() allows to specify default value if label(key) is not present
- Series.loc() is to access a group by labels(keys)

<h1 style="font-size:60px;font-weight:bold;">Binary Operator</h1>

```
s1, s2 = pd.Series([1,2,3]), pd.Series([4,5,6])
```
```
s1.add(s2)
s1.sub(s2)
s1.mul(s2)
s1.div(s2)
```

```
s1.truediv(s2)
s1.floordiv(s2)
s1.mod(s2)
```


```
s1.pow(3)
s1.pow(s2)
```

```
s1.radd(1)
s1.rsub(1)
s1.rmul(2)
s1.rdiv(2)
s1.rfloordiv(2)
s1.rmod(2)
s1.rpow(2)
```
