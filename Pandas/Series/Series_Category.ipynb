__Series.cat.categories__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.categories
```

__Series.cat.ordered__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.ordered
```

__Series.cat.codes__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.codes
```

__Series.cat.rename_categories(*args, **kwargs)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.rename_categories(['a', 'b', 'c'])
```

__Series.cat.reorder_categories(*args, **kwargs)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.reorder_categories(['C', 'B', 'A'], ordered=True)
```

__Series.cat.add_categories(*args, **kwargs)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.add_categories(['D', 'E'])
```

__Series.cat.remove_categories(*args, **kwargs)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.remove_categories(['A'])
```

__Series.cat.remove_unused_categories(*args, ...)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.remove_unused_categories()
```

__Series.cat.set_categories(*args, **kwargs)__
```
s = pd.Series(['A', 'B', 'C', 'A'], dtype='category')
s.cat.set_categories(['C', 'B', 'A'])
```
