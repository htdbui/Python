<h1 style="font-size:60px;font-weight:bold;">Plotting</h1>

__Series.plot([kind, ax, figsize, ...])__
```
s = pd.Series(np.random.randn(1000))
s.plot(kind='hist')
```

__Series.plot.area([x, y, stacked])__
```
df = pd.DataFrame({'A': np.random.randn(1000), 'B': np.random.randn(1000)+1})
df.plot.area(stacked=False)
```

__Series.plot.bar([x, y])__
```
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.bar()
```

__Series.plot.barh([x, y])__
```
df = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df.plot.barh()
```

__Series.plot.box([by])__
```
df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
df.plot.box()
```
__Series.plot.density([bw_method, ind])__
```
s = pd.Series(np.random.randn(1000))
s.plot(kind='density')
```

__Series.plot.hist([by, bins])__
```
s = pd.Series(np.random.randn(1000))
s.plot(kind='hist', bins=50)
```

__Series.plot.kde([bw_method, ind])__
```
s = pd.Series(np.random.randn(1000))
s.plot(kind='kde')
```

__Series.plot.line([x, y])__
```
s = pd.Series(np.random.randn(1000))
s.plot(kind='line')
```

__Series.plot.pie(**kwargs)__
```
s = pd.Series([20, 30, 40, 10], index=['A', 'B', 'C', 'D'], name='pie')
s.plot(kind='pie', figsize=(6, 6))
```

__Series.hist([by, ax, grid, xlabelsize, ...])__
```
s = pd.Series(np.random.randn(1000))
s.hist(bins=50)
```
