<h1 style="font-size:60px;font-weight:bold;">Time Series-related</h1>

__Series.asfreq(freq[, method, how, ...])__: return a new Series
```
s = pd.Series([1, 2, 3], index=pd.date_range(start='2022-01-01', periods=3, freq='D'))
s.asfreq('H', method='pad')
```
__Series.asof(where[, subset])__: return a new Series
```
s = pd.Series([1, 3, 2, 4], index=[0, 1, 3, 4])
s.asof(2)
# explain: till 2 there are three elements
```
__Series.shift([periods, freq, axis, fill_value])__: return a new Series
```
s = pd.Series([1, 2, 3])
s.shift(1)
```
__Series.first_valid_index() and last_valid_index()__: return the index of the first (last) non-null value
```
s = pd.Series([np.nan, 2, 3, np.nan])
s.first_valid_index()
s.last_valid_index()
```
__Series.resample(rule[, axis, closed, label, ...])__: return a new resampled Series
```
s = pd.Series(np.random.randn(1000), index=pd.date_range(start='2022-01-01', periods=1000))
s.resample('M').mean()
```
__Series.tz_convert(tz[, axis, level, copy])__: return a new Series with converted time zone
```
s = pd.Series([1, 2, 3], index=pd.date_range(start='2022-01-01', periods=3, tz='UTC'))
s.tz_convert('US/Eastern')
```
__Series.tz_localize(tz[, axis, level, copy, ...])__: return a new Series with localized time zone
```
s = pd.Series([1, 2, 3], index=pd.date_range(start='2022-01-01', periods=3))
s.tz_localize('US/Eastern')
```
__Series.at_time(time[, asof, axis])__: return a new Series with values at specified time
```
s = pd.Series([1, 2, 3], index=pd.date_range(start='2022-01-01', periods=3, freq='H'))
s.at_time('02:00')
```
__Series.between_time(start_time, end_time[, ...])__: return a new Series bebetween start and end times
```
s = pd.Series([1, 2, 3], index=pd.date_range(start='2022-01-01', periods=3, freq='H'))
s.between_time('02:00', '04:00')
```
