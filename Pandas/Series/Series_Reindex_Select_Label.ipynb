<h1 style="font-size:60px;font-weight:bold;">Reindexing / Selection / Label Manipulation</h1>

__Series.align(other[, join, axis, level, ...])__
```
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([4, 5, 6], index=['b', 'c', 'd'])
s1.align(s2, fill_value=0)
# explain: s1_aligned, s2_aligned
```
__Series.drop([labels, axis, index, columns, ...])__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.drop(['b', 'c'])
```
__Series.droplevel(level[, axis])__
```
s = pd.Series([1, 2, 3], index=[['a', 'a', 'b'], ['x', 'y', 'z']])
s.droplevel(0)
# level at position 0 is dropped
```
__Series.drop_duplicates(*[, keep, inplace, ...])__
```
s = pd.Series([1, 2, 2, 3, 3, 3])
s.drop_duplicates()
```
__Series.duplicated([keep])__
```
s = pd.Series([1, 2, 2, 3, 3, 3])
s.duplicated()
```
__Series.equals(other)__
```
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s3 = pd.Series([1, 2, 3], index=['a', 'b', 'd'])
s1.equals(s2)
s1.equals(s3)
```
__Series.first_valid_index()__: return the index of the first non-missing value
```
s = pd.Series([np.nan, np.nan, 1, 2, 3])
s.first_valid_index()
```
__Series.head([n])__: default is 5
```
s = pd.Series([1, 2, 3, 4, 5])
s.head(3)
```
__Series.idxmax([axis, skipna]) and Series.idxmin([axis, skipna])__
```
s = pd.Series([1, 2, 3, 4, 5])
s.idxmax()
s.idxmin()
```
__Series.isin(values)__
```
s = pd.Series([1, 2, 3, 4, 5])
s.isin([2, 4, 6])
```
__Series.last_valid_index()__
```
s = pd.Series([1, 2, 3, np.nan, np.nan])
s.last_valid_index()
```
__Series.reindex([index, axis, method, copy, ...])__: return a new Series
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.reindex(['a', 'b', 'd'], fill_value=0)
```
__Series.reindex_like(other[, method, copy, ...])__: return a new Series
```
s1 = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s2 = pd.Series([4, 5, 6], index=['b', 'c', 'd'])
s1.reindex_like(s2)
```
__Series.rename([index, axis, copy, inplace, ...])__: return a new Series
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.rename({'a': 'x', 'b': 'y', 'c': 'z'})
```
__Series.rename_axis([mapper, index, axis, ...])__: return a new Series
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.rename_axis('letters')
```
__Series.reset_index([level, drop, name, ...])__: return a view
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.reset_index()
```
__Series.take(indices[, axis])__: return a new Series
```
s = pd.Series([1, 2, 3, 4, 5])
s.take([1, 3])
```
__Series.tail([n])__
```
s = pd.Series([1, 2, 3, 4, 5])
s.tail(3)
```
__Series.truncate([before, after, axis, copy])__: return a new Series
```
s = pd.Series([1, 2, 3, 4, 5], index=['a', 'b', 'c', 'd', 'e'])
s.truncate(before='c', after='d')
```
__Series.where(cond[, other, inplace, axis, level])__: return a new Series
```
s = pd.Series([1, 2, 3, 4, 5])
s.where(s > 3, 0)
```
__Series.mask(cond[, other, inplace, axis, level])__: return a new Series
```
s = pd.Series([1, 2, 3, 4, 5])
s.mask(s > 3, -1)
```
__Series.add_prefix(prefix[, axis]) and suffix__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.add_prefix('x_')
s.add_suffix('_y')
```
__Series.filter([items, like, regex, axis])__
```
s = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
s.filter(items=['a', 'c'])
```
