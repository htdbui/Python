```
s = pd.Series([1, 2, 3, 4, 5], index=pd.date_range('2022-01-01', periods=5, freq='D'))
r = s.resample('W')
for key, group in r:
    print(key, group)
```

<h1 style="font-size:60px;font-weight:bold;">Function Application</h1>


__Resampler.apply([func])__

```
s = pd.Series([1, 2, 3, 4, 5], index=pd.date_range('2022-01-01', periods=5, freq='D'))
r = s.resample('W')
r.apply(lambda x: x.mean() - x.std())
```

__Resampler.aggregate([func])__

```
s = pd.Series([1, 2, 3, 4, 5], index=pd.date_range('2022-01-01', periods=5, freq='D'))
r = s.resample('W')
r.aggregate(['sum', 'mean'])
```

__Resampler.transform(arg, *args, **kwargs)__

```
s = pd.Series([1, 2, 3, 4, 5], index=pd.date_range('2022-01-01', periods=5, freq='D'))
r = s.resample('W')
r.transform(lambda x: (x - x.mean()) / x.std())
```

__Resampler.pipe(func, *args, **kwargs)__

```
s = pd.Series([1, 2, 3, 4, 5], index=pd.date_range('2022-01-01', periods=5, freq='D'))
r = s.resample('W')
r.pipe(lambda x: x / x.sum())
```
