<h1 style="font-size:60px;font-weight:bold;">Seeding and State</h1>

    - get_state returns a tuple that contains the current state of the random number generator
    - set_state sets the state of the random number generator to a given tuple
    - seed sets the seed of the random number generator to a given integer or array
```
np.random.get_state()
state = np.random.get_state()
np.random.rand(3)
np.random.set_state(state)
np.random.rand(3)
---
np.random.seed(123)
```

<h1 style="font-size:60px;font-weight:bold;">Simple Random Data</h1>

# - rand
    uniform distributed [0, 1)
```
np.random.rand(1)
np.random.rand(2, 3)
```
# - randn
    standard normal distributed
```
np.random.randn(2, 4)
```
# - randint
    syntax(low, high=None, size=None, dtype=int)
```
# discrete uniform distributed [0, low)
np.random.randint(10)
# discrete uniform distributed [low, high)
np.random.randint(1, 20, (3, 5))
```
# - random_integers
    - similar to np.random.randint, except that it includes both endpoints of the interval [low, high].
    - this function is deprecated because it includes the endpoint.
    - randint behaves consistent with np.arange. Because of that, it is preferred.
# - random_sample
    - uniform distributed [0, 1) 
    - equivalent to np.random.rand, just different from syntax
    - np.random.random_sample(size=None)
    - np.random.rand(d0, d1, ..., dn)
```
np.random.random_sample()
np.random.random_sample((4,4))
```
# - choice
```
np.random.choice(['a', 'b', 'c', 'd', 'e'])
np.random.choice(['a', 'b', 'c', 'd', 'e'], size=(2, 2), p=[0.1, 0.2, 0.3, 0.4, 0])
np.random.choice(['a', 'b', 'c', 'd', 'e'], size=(2, 2), p=[0.1, 0.2, 0.3, 0.4, 0], replace=False)
---
# discrete uniform distributed [0,a)
np.random.choice(9, size=(3, 3))
```

<h1 style="font-size:60px;font-weight:bold;">Permutations</h1>

# - shuffle
    return None and change the order of array
```
a = [1,2,3,4,5]
np.random.shuffle(a) # order of a changed
a
```
# - permutation
    return a new ndarray
```
a = [1,2,3,4,5]
np.random.permutation(a) # order of a unchanged
---
b = np.random.permutation(a)
a[0] = 0; b
```
