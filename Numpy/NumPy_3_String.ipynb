# - Creat a String
```
np.str_('abc')
np.str_(['a', 'b'])
```

# - Methods
## + Addition
```
np.char.add('abc', 'cde')
```
## + Multiplication
```
np.char.multiply('abC', 3)
```
## + Capitalization
``` 
np.char.lower('ABC')
np.char.upper('abc')
np.char.capitalize('abc def')
np.char.title('abc def')
np.char.swapcase('Abc')
```
## + Format Space
```
np.char.center('abc', 5, '_')
np.char.rjust('abc', 5, '_')
np.char.ljust('abc', 5, '_')
np.char.expandtabs('Geeks\tFor\tGeeks')
```
## + Joining
```
np.char.join('*','||')
np.char.join(['*','+'],['//','||'])
```
## + Splitting
```
a = np.array(['a,b,c', 'd,e,f'])
np.char.split(a, sep=',')
np.char.rsplit(a, sep=',')
np.char.split(a, sep=',', maxsplit=1)
np.char.rsplit(a, sep=',', maxsplit=1)
---
np.char.splitlines('hello\nhow are you?')
np.char.splitlines('hello\nhow are you?', keepends=True)
---
np.char.partition(a, sep=',')
np.char.rpartition(a, sep=',')
```
## + Strip
```
np.char.lstrip( np.array(['  hello', 'world  ', '  python ']) )
np.char.rstrip( np.array(['  hello', 'world  ', '  python']) )
np.char.strip( np.array(['  hello', 'world  ', '  python']) )
```
## + replace, translate, zfill
```
np.char.replace('abc', old='b', new='B')
np.char.translate('abc', table=str.maketrans('ac', '<>'))
np.char.zfill('123', width=5)
```

# - Comparision
```
a = np.array(['hello', 'world', 'numpy'])
b = np.array(['HELLO', 'world ', 'NumPy'])
```
## > >=
```
a > b
np.char.greater(a,b)
a >= b
np.char.greater_equal(a,b)
```
## < <=
```
a < b
np.char.less(a,b)
a <= b
np.char.less_equal(a,b)
```
## == !=
```
a == b
np.char.equal(a,b)
a != b
np.char.equal_not(a,b)
```
## + compare_chararrays
```
np.char.compare_chararrays(a, b, cmp='==', rstrip=True) # equal
np.char.compare_chararrays(a, b, cmp='!=', rstrip=True) # not equal
np.char.compare_chararrays(a, b, cmp='<', rstrip=True) # less
```

# - String Information
    a = np.array(['hello', 'world', 'numpy'])
## + endswith, startswith
```

np.char.count(a, sub='l')
np.char.count(a, sub='o', start=2)
```

```
np.char.find(a, sub='l')
np.char.rfind(a, sub='l')
np.char.find(a, sub='o', start=2)
np.char.find(a, sub='o', end=4)
```
## + isalpha, isalnum, isdecimal, isnumeric, isdigit
```
a = np.array(['abc', '123', 'a1b2', 'Ⅸ', '½'])
np.char.isalpha(a) 
np.char.isalnum(a) # alpha + number
np.char.isdecimal(a)
np.char.isnumeric(a)
np.char.isdigit(a)
```
## + islower, isupper, istitle, isspace
